```python
age_days =[]
for i, row in users.iterrows():
  age_days.append(row['Age']*365)
```

```python
users2=data.loc[:,('CustomerId','Surname','Geography','Age','Gender','EstimatedSalary')]
users2.head()
 
```

```python
users2['Age_in_days']=age_days
users2.head()
```

```python
def age_to_days(x):
  return x*365
users2['Age_d3']=users2['Age'].apply(age_to_days)
users2.head()
```

```python
import time
from tqdm import tqdm
tqdm.pandas()
def age_to_days2(x):
  time.sleep(0.001)
  return x*365
users2['Age']=users2['Age'].progress_apply(age_to_days2)
```
