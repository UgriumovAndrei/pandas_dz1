```python
def weight(x):
  return float(x[:-2])
df['Weight1']=df['Weight'].apply(weight)
```

```python
df.drop(columns='Weight', inplace=True)
df.head(5)
```

```python
def Memory_type(x):
  temp=[i.lstrip() for i in x.split(' ')]

  return ' '.join(temp[1:])

Memory_type('1TB HDD')
```

```python
def memory_amount(x):
  temp=x.split(' ')
  return temp[0][:-2]
memory_amount('1TB HDD')
```

```python
df['memory_type']=df['Memory'].apply(Memory_type)
df['memory_amount']=df['Memory'].apply(memory_amount)
```

```python
df.head(2)
```

```python
def cpu_company(x):
  temp=x.split(' ')
  return temp[0]
cpu_company('Intel Iris Plus Graphics 640')
```

```python
plt.figure(figsize=(12,6))
plt.hist(df['cpu_company'], bins=5)
plt.title('Распределение по комании-производителю')
plt.xlabel('Компания')
plt.ylabel('кол-во')
plt.xticks(rotation=80);
```

```python
 cpucomps=df[['cpu_company','Company']].value_counts().unstack('cpu_company').fillna(0)[['AMD','Intel']].reset_index()
 cpucomps
```

```python
plt.figure(figsize=(12,6))
n_ticks = np.arange(len(cpucomps['AMD']))
offset=0.2
w=0.4
plt.bar(n_ticks - offset, cpucomps['Intel'],width=w)
plt.bar(n_ticks + offset, cpucomps['AMD'],width=w)

plt.title('Комании-ноутбуки и компании-процессоры')
plt.xlabel('Компания')
plt.ylabel('Кол-во')
plt.legend(['Intel', 'AMD'])
plt.xticks(n_ticks, cpucomps['Company'], rotation=30);
```

```python
 cpucomps2=df[['cpu_company','Company']].value_counts(normalize=True).unstack('cpu_company').fillna(0)[['AMD','Intel']].reset_index()
 cpucomps2
```

```python
plt.figure(figsize=(12,6))
n_ticks = np.arange(len(cpucomps2['AMD']))
offset=0.2
w=0.4
plt.bar(n_ticks , cpucomps2['Intel'],width=w)
plt.bar(n_ticks , cpucomps2['AMD'],width=w, bottom=cpucomps2['Intel'])

plt.title('Комании-ноутбуки и компании-процессоры')
plt.xlabel('Компания')
plt.ylabel('Кол-во')
plt.legend(['Intel', 'AMD'])
plt.xticks(n_ticks, cpucomps2['Company'], rotation=30);
```
