Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix=df.corr()
corr_matrix=np.round(corr_matrix,1)
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Corr_matrx')
```

для нахождения корреляций, построена матрица корреляций и тепловая карта , подсвечивающая самые яркие зависимости
Выводы:по результатам, самые сильные зависимости выявлены по категориям sqft_living,grade, а также sqft_above . Чтобы убедится в этом построим для них отдельные графики.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Данный график иллюстрирует зависимость котегорий цены (price) и grade - оценки. Это может быть взято на вооружение маркетологами, которые повышая оценку домов пользователями(например сайтов оценки недвижимости) будут повышать их цену.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Данные график также показывает прямую взаимосвязь стоимости и жилой площади дома. Очевидно, что чем больше площпдь - тем выше цена, ведь как правило продают именно квадратные мтеры.

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg');
```

Наконец, наиболее интересный график. В хлде анализа данных было выявлено, что в тройку также входит зависимость цены от возвышения дома. Это достаточно интересная корелляция, которая не сразу бросается в глаза. Можно предположить, что возвышение влияет на плотность населения района, где расположен дом,  вид из окна, а также экологию.
